```python
def calculate_due_date(transaction_date):
    today = datetime.date.today()
    if transaction_date.weekday() >= 1:
        if today.year == transaction_date.year and datetime.date.isocalendar(today)[1] == datetime.date.isocalendar(transaction_date + datetime.timedelta(days=1))[1]:
            return 1
    return "Missing Due Date"

for r in range(len(df["Transaction Date"])):
    test1=datetime.datetime.strptime(df["Transaction Date"][i], "%Y-%m-%d")
    t1r=calculate_due_date(test1)
    cdf["Current Week"][i]=t1r
```

```python
def Current_Month(transaction_date):
    today = datetime.date.today()
    if transaction_date.month == today.month and transaction_date.year == today.year:
        return 1
    else:
        return "Missing Date"
for r in range(len(df["Transaction Date"])):
    test1=datetime.datetime.strptime(df["Transaction Date"][i], "%Y-%m-%d")
    t1r=Current_Month(test1)
    cdf["Current Month"][i]=t1r    
```

```python
def Current_Year(transaction_date):
    today = datetime.datetime.now()
    if transaction_date.year == today.year:
        return 1
    else:
        return 0
for r in range(len(df["Transaction Date"])):
    test1=datetime.datetime.strptime(df["Transaction Date"][i], "%Y-%m-%d")
    t1r=Current_Year(test1)
    cdf["Current Year"][i]=t1r     
```

```python
def calculate_week_number(date):
    return date.isocalendar()[1]
def Previous_Week(transaction_date):
    today = datetime.date.today()    
    if transaction_date.weekday() >= 1:
        if today.year == transaction_date.year and calculate_week_number(today) - 1 == calculate_week_number(transaction_date + datetime.timedelta(days=1)):
            return 1            
    return "Missing Date"
for r in range(len(df["Transaction Date"])):
    test1=datetime.datetime.strptime(df["Transaction Date"][i], "%Y-%m-%d")
    t1r=Previous_Week(test1)
    cdf["Previous Week"][i]=t1r 
```

```python
def is_previous_month(date):
    today = datetime.datetime.today().replace(day=1)  # Get the first day of the current month
    previous_month_end = today - timedelta(days=1)  # Get the last day of the previous month
    
    return date.month < previous_month_end.month or (date.month == previous_month_end.month and date.year < previous_month_end.year)
for r in range(len(df["Transaction Date"])):
    test1=datetime.datetime.strptime(df["Transaction Date"][i], "%Y-%m-%d")
    t1r=is_previous_month(test1)
    cdf["Previous Month"][i]=t1r 
```
